```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[audio_features])
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=5,random_state=42)
km.fit(X_scaled)



```

```python
km.cluster_centers_
```

```python
df["cluster_group"] = km.labels_
df
```

```python
df["cluster_group"].value_counts()
```

```python
df = df.reset_index(drop=True)
```

```python
centers=km.cluster_centers_
distances=[]
for i in range(len(X_scaled)):
  cluster=df.loc[i,"cluster_group"]
  center=centers[cluster]
  distance=np.linalg.norm(X_scaled[i]-center)
  distances.append(distance)
df["distance_from_cluster_center"]=distances
print("Distance column added successfully")
```

```python
def recommend_songs(song_name, df, n=10):

    if song_name not in df['track_name'].values:
        return "Song not found in dataset"


    song_cluster = df.loc[
        df['track_name'] == song_name,
        'cluster_group'
    ].iloc[0]


    recommendations = df[
        (df['cluster_group'] == song_cluster) &
        (df['track_name'] != song_name)
    ]


    recommendations = recommendations.sort_values(
        by='distance_from_cluster_center'
    )

    return recommendations[
        ['track_name','track_artist','playlist_genre']
    ].head(n)
```

```python
recommend_songs("Believer", df)
```

```python
def recommend_by_artist(artist_name, df, n=10):

    if artist_name not in df['track_artist'].values:
        return "Artist not found in dataset"

    recommendations = df[
        df['track_artist'] == artist_name
    ][['track_name','track_artist','playlist_genre']]

    return recommendations.head(n)
```

```python
recommend_by_artist("Ed Sheeran", df)
```

```python
def recommend_by_genre(genre_name, df, n=10):

    if genre_name not in df['playlist_genre'].values:
        return "Genre not found in dataset"

    recommendations = df[
        df['playlist_genre'] == genre_name
    ][['track_name','track_artist','playlist_genre']]

    return recommendations.head(n)
```

```python
recommend_by_genre("pop", df)
```
